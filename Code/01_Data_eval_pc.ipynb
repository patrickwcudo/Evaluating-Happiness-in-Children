## Model with just variables explored from goolge sheet.

```python
# create file with list of columns explored
test_df = df[["AGELAST",
 "RTHLTH31",
 "MNHLTH31",
 "HIBPDX",
 "PRVEV17",
 "TRIEV17",
 "MCREV17",
 "MCDEV17",
 "OPAEV17",
 "UNINS17",
 "INSCOV17",
 "INSURC17",
 "MDUNAB42",
 "TOTEXP17"]]
```

```python
# create ols model with test_df
# imports
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
X = test_df.drop(columns='TOTEXP17')
# add constant
X = sm.add_constant(X, prepend=True)
# set y
y = test_df['TOTEXP17']
```

```python
results = sm.OLS(y, X).fit()
```

```python
results.summary()
```

## This model is everything except columns that are not numbers, so drop_list includes all variables that are objects.
